#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import numpy as np
```

```python
# cargar directo de reopsitorio github
url = 'https://raw.githubusercontent.com/Malcuthrad/Challenge-Telecom-X/refs/heads/main/TelecomX_Data.json'
df = pd.read_json(url)
df.head()
```

#ðŸ”§ TransformaciÃ³n

```python
# normalizar columnas con datos anidados
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# sacar columnas anidadas originales
df.drop(columns=['customer', 'phone', 'internet', 'account'], inplace=True)

# Concatenar el DataFrame original con los nuevos datos normalizados
df = pd.concat([df, customer_df, phone_df, internet_df, account_df], axis=1)

df.head()
```

```python
df.info()
```

```python
# verificar columnas
df.columns
```

```python
# Verificar datos nulos
df.isnull().sum()
```

```python
# Verificar datos duplicados
df.duplicated().sum()
```

```python
df.sample(10)
```

```python
# Reemplazar strings vacios con nan
df = df.replace('', np.nan)
```

```python
df = df.replace(' ', np.nan)
```

```python
# Volver a verificar nulos
df.isnull().sum()
```

```python
# Mostrar solo suma de columnas con datos nulor
df[['Churn', 'Charges.Total']].isnull().sum()
```

```python
# Datos con nan no se estiman relevantes por lo que se eliminan
df = df.dropna(subset=['Churn', 'Charges.Total'])
df[['Churn', 'Charges.Total']].isnull().sum()
```

```python
df.info()
```

```python
# Cambiar Charges.Total de object a float
df['Charges.Total'] = df['Charges.Total'].astype(np.float64)
df.info()
```

```python
# dejar todo en minusculas y eliminar caracteres especiales a columna Contract
df['Contract'] = df['Contract'].str.lower()
df['Contract'] = df['Contract'].str.replace(r'\-', ' ', regex=True).str.strip()
```

```python
df['Contract'].sample(4)
```

```python
# cambiar colunmas con 1 y 0 a yes y no
#colunmas_convertidas = ['Churn', 'SeniorCitizen',	'Partner', 'Dependents', 'PaperlessBilling']
colunma_convertida = ['SeniorCitizen']
df[colunma_convertida] = df[colunma_convertida].replace({1 : 'Yes', 0 : 'No'})
df.sample(4)
```
